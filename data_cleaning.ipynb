```python
from bs4 import BeautifulSoup

def clean_html(text):
    if pd.isna(text):
        return ''
    return BeautifulSoup(text, "html.parser").get_text(separator=' ')
```

```python
df['description'] = df['description'].apply(clean_html)
```
